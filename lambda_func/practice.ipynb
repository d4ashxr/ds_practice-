```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("clean_data.csv")
df
```

Добавление нового признака через лямбда функцию

```python
df["price in rub"] = df.apply(lambda x: x.price * 97, axis=1)
df.head()
```

```python
nan_count = df['region'].isna().sum()
nan_count
```

```python
df["region"] = df.apply(lambda x: x.region_url.split(".")[0][8:], axis=1)
df.head()
```

обьединение строк двух колонок датафрейма

```python
df["mark and model"] = df.apply(lambda x: ": ".join([str(x.manufacturer), str(x.model)]), axis=1)
df.head()
```

заполнение пропусков с помощью лямбда функции

```python
df.fuel.isna().sum()
```

```python
fuel_mode = df.fuel.mode()[0]
df['fuel'] = df['fuel'].apply(lambda x: fuel_mode if pd.isna(x) else x)
```

```python
df.fuel.isna().sum()
```

Преобразование данных в матрицу удобную для машинного обучения

```python
ohe = OneHotEncoder(sparse=False)
ohe.fit(df[["fuel"]])
```

```python
ohe_fuel = ohe.transform(df[["fuel"]])
ohe_fuel
ohe.categories_
```

```python
ohe.inverse_transform(ohe_fuel)
```

```python
df[ohe.get_feature_names_out()] = ohe_fuel
```

```python
df.head()
```

```python
ohe1 = OneHotEncoder(sparse=False)
ohe1
```

```python
ohe1.fit(df[["price_category"]])
ohe_price = ohe1.transform(df[["price_category"]])
```

```python
df[ohe1.get_feature_names_out()] = ohe_price
```

```python
df.head()
```

Нормализация значений для машинного обучения

```python
scaler = StandardScaler()
scaler
```

```python
scaler.fit(df[["year", "price"]])
print("mean:", scaler.mean_, " dispersia:", scaler.var_)
```

```python
scaler_year = scaler.transform(df[["year", "price"]])
```

```python
df[["std_scaler_year", "std_scaler_price"]] = scaler_year
```

```python
import matplotlib.pyplot as plt
```

```python
fig, (axl, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(20, 12))
axl.scatter(df.price, df.year)
axl.set_title("Mexonuag")
axl.set_ylabel('year')
axl.set_xlabel('Стоимость автомобиля')
ax2.scatter(df.std_scaler_price, df.std_scaler_year)
ax2.set_title('После стандартизации')
ax2.set_ylabel('year')
ax2.set_xlabel('Стоимость автомобиля')
fig.suptitle( 'Стоимость автомобиля vs year') ;
```
