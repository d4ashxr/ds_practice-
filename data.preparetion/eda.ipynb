```python
clean_data = data.copy()
clean_data
```

```python
year_avg = clean_data.year.mean()
clean_data.year = clean_data.year.fillna(year_avg)
clean_data.year.isna().sum()
```

```python
clean_data = clean_data.drop(columns=['county', 'size'])
```

```python
clean_data.model = clean_data.model.fillna('other')
```

```python
fuel_mode = clean_data.fuel.mode()
```

```python
clean_data.fuel = clean_data.fuel.fillna(fuel_mode)
```

```python
clean_data.title_status.unique()
```

```python
clean_data.title_status = clean_data.title_status.fillna('not stated')
```

```python
clean_data.year
```

```python
clean_data.year = clean_data.year.astype(int)
clean_data.year.dtype
```

```python
msno.matrix(clean_data)
```

```python
def outliers(data_seria):
    q25 = data_seria.quantile(0.25)
    q75 = data_seria.quantile(0.75)
    dif = q75 - q25
    boundaries = (q25 - 1.5 * dif, q75 + 1.5 * dif)
    return boundaries
```

```python
borders = outliers(clean_data.year)
```

```python
ejections = (clean_data.year > borders[1]) | (clean_data.year < borders[0])
ejections.sum()
```
